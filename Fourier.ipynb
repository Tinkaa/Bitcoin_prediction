```python
import numpy as np
import pandas as pd
import seaborn
from matplotlib import pyplot as plt
```

```python
DF = pd.read_csv("bitcoin_start-240418.csv")
DF = DF[:-2]
prices = np.array(DF["Close Price"])
```

```python
prices
```

```python


# Nyquist Sampling
N = prices.size
T = 1/N # inverse of the sampling rate
x = np.linspace(0.0, 1.0/(2.0*T), int(N/2))

fft = np.fft.rfft(prices)
pwr = np.abs(fft)  # AMPLITUDE
pwr = 2/N * np.abs(pwr[0:np.int(N/2)]) # positive freqs only

#timestep = 1 / fft.size
#freq = np.fft.fftfreq(fft.size, timestep)
print(x.shape)
print(pwr.shape)

plt.plot(x, pwr)
plt.xlim((0, 200))

print(pwr[7])

plt.show()
```

```python
toy = np.zeros(35)
toy[6] = pwr[6]
inverse = np.fft.ifft(toy)
```

```python
frame = np.linspace(0, len(toy)-1, len(toy))

plt.plot(frame, inverse)
plt.show()
```
