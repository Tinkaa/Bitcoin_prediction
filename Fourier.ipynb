```python
import numpy as np
import pandas as pd
import seaborn
from matplotlib import pyplot as plt
```

```python
DF = pd.read_csv("bitcoin_start-240418.csv")
DF = DF[:-2]
prices = np.array(DF["Close Price"])
```

```python
prices
```

```python
fft = np.fft.rfft(prices)
pwr = np.abs(fft)**2

timestep = 1 / fft.size
freq = np.fft.fftfreq(fft.size, timestep)
print(freq)

plt.plot(pwr)
plt.xlim((0, 35))

print(pwr[7])

plt.show()
```

```python
toy = np.zeros(35)
toy[6] = pwr[6]
inverse = np.fft.ifft(toy)
```

```python
frame = np.linspace(0, len(toy)-1, len(toy))

plt.plot(frame, inverse)
plt.show()
```
